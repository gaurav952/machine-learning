```python
import pandas as pd
import numpy as np
from sklearn import *
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions

%matplotlib inline
```

###### The dataset used is the famous IRIS dataset to find the complex boundaries

```python
df = pd.read_csv("../Data/iris.csv")
df.head()
```

```python
features = ["SepalLengthCm", "PetalLengthCm"]
```

```python
df.Species.value_counts()
```

```python
_, ax = plt.subplots()
colors = ["red", "green", "blue"]

df[df.Species == "Iris-setosa"].plot.scatter(features[0], 
                    features[1], ax = ax, color = colors[0], label = "Iris-setosa")
df[df.Species == "Iris-virginica"].plot.scatter(features[0], 
                    features[1], ax = ax, color = colors[1], label = "Iris-virginica")
df[df.Species == "Iris-versicolor"].plot.scatter(features[0], 
                    features[1], ax = ax, color = colors[2], label = "Iris-versicolor")
```

```python
target = "default"
y = np.where(df.Species == "Iris-setosa", 1, 0) #outcome variable
X = df[features]
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y,
                        test_size = 0.3, random_state = 1234)
X_train.shape, X_test.shape

pipe = pipeline.Pipeline([
    ("poly", preprocessing.PolynomialFeatures(degree=1, include_bias=False)),
    ("scaler", preprocessing.StandardScaler()),
    ("est", tree.DecisionTreeClassifier(max_depth=4, criterion="entropy", random_state=1))
])

pipe.fit(X_train, y_train)
print("training: ", pipe.score(X_train, y_train),
"\ntesting: ", pipe.score(X_test, y_test))
plot_decision_regions(X.values, y, pipe)
plt.xlabel(features[0])
plt.ylabel(features[1])
plt.title("Decision region")
```

```python
y = np.where(df.Species == "Iris-versicolor", 1, 0)
X = df[features]
X = (X - X.mean(axis = 0))/X.std(axis = 0)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.3, random_state = 1)
X.shape, y_train.shape, y_test.shape, X_test.shape
```

```python
import keras
import tensorflow as tf
```

```python
tf.set_random_seed(1)
np.random.seed(1)

def build_model():
    adam = keras.optimizers.Adam(lr=0.01)
    model = keras.Sequential()
    model.add(keras.layers.Dense(units=3, activation="relu", input_shape=(2,)))
    model.add(keras.layers.Dense(units=3, activation="relu"))
    
    model.add(keras.layers.Dense(units=1, activation="sigmoid"))
    model.compile(loss="binary_crossentropy", metrics=["accuracy"], optimizer=adam) 
    return model

model = build_model()
model.summary()
model.fit(X_train, y_train, batch_size=5, epochs=50)
model.evaluate(X_test, y_test, batch_size=10)
```

```python
plot_decision_regions(X.values, y, model)
```
